## visualising

```python
rnd_spend = df.iloc[:, 0]
profit = df.iloc[:, -1]

plt.scatter(rnd_spend, profit)
plt.title('profit vs RND spend')
plt.xlabel('misc spend')
plt.ylabel('profit')
plt.show()
```

```python
admin_spend = df.iloc[:, 1]
profit = df.iloc[:, -1]

plt.scatter(admin_spend, profit)
plt.title('profit vs admin spend')
plt.xlabel('admin spend')
plt.ylabel('profit')
plt.show()
```

```python
marketting_spend = df.iloc[:, 2]
profit = df.iloc[:, -1]

plt.scatter(marketting_spend, profit)
plt.title('profit vs marketting spend')
plt.xlabel('marketting spend')
plt.ylabel('profit')
plt.show()
```

```python
state = df.iloc[:, 3]
profit = df.iloc[:, -1]

plt.scatter(state, profit)
plt.title('profit vs state')
plt.xlabel('state')
plt.ylabel('profit')
plt.show()
```
