# Data Visualisation

## importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## importing dataset

```python
df = pd.read_csv("data.csv")
df
```

## visualising

```python
misc_spend = df.iloc[:, 0]
profit = df.iloc[:, -1]

plt.scatter(misc_spend, profit)
plt.title('profit vs misc spend')
plt.xlabel('misc spend')
plt.ylabel('profit')
plt.show()
```

```python
admin_spend = df.iloc[:, 1]
profit = df.iloc[:, -1]

plt.scatter(admin_spend, profit)
plt.title('profit vs admin spend')
plt.xlabel('admin spend')
plt.ylabel('profit')
plt.show()
```

```python
marketting_spend = df.iloc[:, 2]
profit = df.iloc[:, -1]

plt.scatter(marketting_spend, profit)
plt.title('profit vs marketting spend')
plt.xlabel('marketting spend')
plt.ylabel('profit')
plt.show()
```

```python
state = df.iloc[:, 3]
profit = df.iloc[:, -1]

plt.scatter(state, profit)
plt.title('profit vs state')
plt.xlabel('state')
plt.ylabel('profit')
plt.show()
```
